```python
from itertools import combinations_with_replacement
from tqdm import tqdm

def encode_string(s):
  if len(s) != 6:
    raise ValueError("String must be exactly 6 characters long")
  return hex(sum(ord(char) for char in s))

def brute_force_keys(n=10):
  valid_sets = []
  alphabet = "abcdefghijklmnopqrstuvwxyz"  # 26 characters
  for comb in tqdm(combinations_with_replacement(alphabet, 6)):
    candidate = "".join(comb)
    if encode_string(candidate) == "0x277":
      valid_sets.append(candidate)
      
  return valid_sets

# Also every possible permutation of the candidates is a valid key
candidates = brute_force_keys()

print("Found candidates:")
candidates_per_line = 15
for i in range(0, len(candidates), candidates_per_line):
  print(candidates[i:i+candidates_per_line])
```
